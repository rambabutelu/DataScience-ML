```python
#importing th enecessary libraries
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import fix_yahoo_finance as yf
yf.pdr_override() 
import datetime
%matplotlib inline
```

```python
'''
We need to get data using pandas datareader. We will get stock information for the following banks:

Bank of America
CitiGroup
Goldman Sachs
JPMorgan Chase
Morgan Stanley
Wells Fargo
'''

#setting the start and end date
start = datetime.datetime(2009, 1, 1)
end = datetime.datetime(2019, 1, 1)
```

```python
#Getting the data using pandas-datareader for different banks

#For Bank of America
BAC = data.get_data_yahoo("BAC", start, end)

#For CitiGroup
CG = data.get_data_yahoo("C", start, end)

#For Goldman Sachs
GS = data.get_data_yahoo("GS", start, end)

#For JPMorgan Chase
JPM = data.get_data_yahoo("JPM", start, end)

#For Morgan Stanley
MS = data.get_data_yahoo("MS", start, end)

#For Wells Fargo
WFC = data.get_data_yahoo("WFC", start, end)
```

```python
#creating a list of ticker symbols
tickers = ['BAC', 'CG', 'GS', 'JPM', 'MS', 'WFC']
```

```python
#concatenating along columns with tickers as keys
bank_stocks = pd.concat([BAC, CG, GS, JPM, MS, WFC], axis=1, keys=tickers)
```

```python
#checking the head of the new concatenated DF
bank_stocks.head()
```

```python
#setting the columns name
bank_stocks.columns.names = ['Bank Ticker','Stock Info']
```

```python
#checking the head
bank_stocks.head()
```

```python
#checking the max Closing price for each bank's stock
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```
